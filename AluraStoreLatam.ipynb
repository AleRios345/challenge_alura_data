### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas = [tienda,tienda2,tienda3, tienda4]

tienda.head()
```

#1. Análisis de facturación



```python
ingresos = []

for i,t in enumerate(tiendas):
    ingreso_total =  t['Precio'].sum()
    ingresos.append(ingreso_total)
    print(f'Tienda {i+1}\nIngreso Total: {ingreso_total}\n')

# Grafico de barras
plt.figure(figsize=(8, 5))
plt.bar([f'Tienda {i+1}' for i in range(len(tiendas))], ingresos, color='mediumseagreen')
plt.title('Ingreso Total por Tienda')
plt.ylabel('Ingreso ($)')
plt.xlabel('Tienda')
plt.tight_layout()
plt.show()

# Grafico Circular
plt.figure(figsize=(7, 7))
plt.pie(ingresos, labels=[f'Tienda {i+1}' for i in range(len(tiendas))], 
        autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Porcentaje de Ingreso Total por Tienda')
plt.axis('equal') 
plt.show()



```

# 2. Ventas por categoría

```python
print("Tienda 1: Ventas por Categoria")
ventas_categoria = tienda.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=True)

plt.figure(figsize=(10, 6))
ventas_categoria.plot(kind='barh', color='skyblue')
plt.title('Ventas por Categoría - Tienda 4')
plt.xlabel('Total de Ventas ($)')
plt.ylabel('Categoría')
plt.tight_layout()
plt.show()
```

```python
print("Tienda 2: Ventas por Categoria")
ventas_categoria = tienda2.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=True)

plt.figure(figsize=(10, 6))
ventas_categoria.plot(kind='barh', color='skyblue')
plt.title('Ventas por Categoría - Tienda 4')
plt.xlabel('Total de Ventas ($)')
plt.ylabel('Categoría')
plt.tight_layout()
plt.show()
```

```python
print("Tienda 3: Ventas por Categoria")
ventas_categoria = tienda3.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=True)

plt.figure(figsize=(10, 6))
ventas_categoria.plot(kind='barh', color='skyblue')
plt.title('Ventas por Categoría - Tienda 4')
plt.xlabel('Total de Ventas ($)')
plt.ylabel('Categoría')
plt.tight_layout()
plt.show()
```

```python
print("Tienda 4: Ventas por Categoria")
ventas_categoria = tienda4.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=True)

plt.figure(figsize=(10, 6))
ventas_categoria.plot(kind='barh', color='skyblue')
plt.title('Ventas por Categoría - Tienda 4')
plt.xlabel('Total de Ventas ($)')
plt.ylabel('Categoría')
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
for i, t in enumerate(tiendas):
    print(f'Calificación Promedio: Tienda {i+1}:\n {t['Calificación'].mean()}')
```

# 4. Productos más y menos vendidos

```python
for i, t in enumerate(tiendas):
    ventas_por_producto = t.groupby('Producto')['Precio'].sum()
    mas_vendido = t.groupby('Producto')['Precio'].sum().idxmax()
    menos_vendido =  t.groupby('Producto')['Precio'].sum().idxmin()
    total_ventas_menor = t.groupby('Producto')['Precio'].sum().min()
    total_ventas = t.groupby('Producto')['Precio'].sum().max()
    print(f'Tienda {i+1}:\nProducto mas vendido: {mas_vendido}\nTotal de ventas: {total_ventas}\nProducto menos vendido: {menos_vendido}\nTotal de ventas: {total_ventas_menor}')
    
    plt.figure(figsize=(10, 6))
    ventas_por_producto.plot(kind='bar', color='skyblue')
    plt.title(f'Ventas por Producto - Tienda {i+1}')
    plt.ylabel('Total de Ventas ($)')
    plt.xlabel('Producto')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
   
```

# 5. Envío promedio por tienda

```python
costos_promedio = []
nombres_tiendas = []

for i, t in enumerate(tiendas):
    promedio = t['Costo de envío'].mean()
    costos_promedio.append(promedio)
    nombres_tiendas.append(f'Tienda {i+1}')
    print(f'Costo Promedio de Envio - Tienda {i+1}:\n {promedio}\n')
    
# Grafica de Barras   
plt.figure(figsize=(8, 5))
plt.bar(nombres_tiendas, costos_promedio, color='lightseagreen')
plt.title('Costo Promedio de Envío por Tienda')
plt.ylabel('Costo Promedio ($)')
plt.xlabel('Tienda')
for i, costo in enumerate(costos_promedio):
    plt.text(i, costo + 0.5, f'{costo:.2f}', ha='center')
plt.tight_layout()
plt.show()

# BoxPlots por tienda
plt.figure(figsize=(10, 6))
plt.boxplot([t['Costo de envío'] for t in tiendas], labels=[f'Tienda {i+1}' for i in range(len(tiendas))])
plt.title('Distribución del Costo de Envío por Tienda')
plt.ylabel('Costo de Envío ($)')
plt.xlabel('Tienda')
plt.grid(True)
plt.tight_layout()
plt.show()
```
